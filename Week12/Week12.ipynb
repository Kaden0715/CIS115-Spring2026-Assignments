```python
# NAME: rectangle.py

# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define calculate_rectangle(length, width)
def calculate_rectangle(length, width):
# 2. Find area
    area = length * width
# 3. Find perimeter
    perimeter = 2 * (length + width)
# 4. Pack area and perimeter into a tuple
    result = (area, perimeter)
# 5. Return the tuple
    return result

# PROGRAM LOGIC
# 1. Call function with (5, 10)
test_result = calculate_rectangle(5, 10)
# 2. Print returned tuple
print(test_result)
# 3. Ask user for length
length = float(input("Enter length: "))
# 4. Ask user for width
width = float(input("Enter width: "))
# 5. Call function and store returned tuple
result = calculate_rectangle(length, width)
# 6. Unpack tuple into my_area and my_perimeter
my_area, my_perimeter = result
# 7. Print results
print("\n--- Results ---")
print("The Area is: " + str(my_area))
print("The Perimeter is: " + str(my_perimeter))
```

```python
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define is_palindrome(word)
def is_palindrome(word):
# 2. Convert word to lowercase
    lower_word = word.lower()
# 3. Reverse word using slicing
    reversed_word = lower_word[::-1]
# 4. Compare lowercase word to reversed word
#    - If equal, return True
    if lower_word == reversed_word:
        return True
#    - Else, return False
    else:
        return False

# PROGRAM LOGIC
# 1. Input for word
word = input("Enter a word: ")
# 2. Call is_palindrome(word)
#    - Store returned value
result = is_palindrome(word)
# 3. If True, print is a palindrome
if result == True:
    print(word + " is a palindrome.")
# 4. Else, print is not a palindrome
else:
    print(word + " is not a palindrome.")
```

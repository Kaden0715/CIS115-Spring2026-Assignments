```python
# NAME: roster.py

# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define add_players(roster)
def add_players(roster):
# 2. Use a while loop to ask for player names
    while True:
        name = input("Enter player name (or 'quit' to stop): ")
# 3. If user enters "quit", stop the loop
        if name == "quit":
            break
# 4. Else, add player name to roster using append()
        else:
            roster.append(name)
# 5. Define is_on_roster(roster, name)
def is_on_roster(roster, name):
# 6. Loop through each player in roster
    for player in roster:
# 7. If name matches a player, return True
        if player == name:
            return True
# 8. After loop ends, return False
    return False

# PROGRAM LOGIC
# 1. Create empty list called roster
roster = []
# 2. Call add_players(roster)
add_players(roster)
# 3. Print roster to test if names were added
print(roster)
# 4. Ask user for name to search
search_name = input("Who would you like to search for? ")
# 5. Call is_on_roster() and store returned value
result = is_on_roster(roster, search_name)
# 6. If True, print "Player found!"
if result == True:
    print("Result: Player found!")
# 7. Else, print "Player not found."
else:
    print("Result: Player not found.")
```

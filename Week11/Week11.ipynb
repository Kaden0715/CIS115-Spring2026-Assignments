```python
# NAME: gradebook.py

# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define add_students(gradebook)
def add_students(gradebook):
# 2. Use a while loop to ask for student name
    while True:
        name = input("Enter student name (or 'done'): ")
# 3. If user enters "done", stop the loop
        if name == "done":
            break
# 4. Else, ask for students grade
        else:
            grade = float(input("Enter " + name + "'s grade: "))
# 5. Add name and grade to dictionary
            gradebook[name] = grade
# 6. Define calculate_average(gradebook)
def calculate_average(gradebook):
# 7. Get all grades using .values()
    grades = gradebook.values()
# 8. Add all grades together
    total = 0
    for g in grades:
        total = total + g
# 9. Divide total by number of students
    count = len(gradebook)
# 10. Return average
# 11. Handle case if no students (avoid division by zero)
    if count == 0:
        return 0
    else:
        average = total / count
        return average

# PROGRAM LOGIC
# 1. Create empty dictionary called gradebook
gradebook = {}
# 2. Call add_students(gradebook)
add_students(gradebook)
# 3. Print gradebook to test if data was added
print(gradebook)
# 4. Call calculate_average() and store returned value
avg = calculate_average(gradebook)
# 5. Print each students name and grade
print("\n--- Class Report ---")
for name in gradebook:
    print(name + ": " + str(gradebook[name]))
# 6. Print class average (formatted to 2 decimal places)
print("Class Average: " + format(avg, ".2f"))
```
